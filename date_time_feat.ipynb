```python
def time_feature(df):
    """
    This function take dataframe as input.
    
    Function caluclate day_of_week, day_of_month and month_of_year
    and return dataframe
    """

    df['day'] = df['DATE'].dt.day
    df['day_of_week'] = df['DATE'].dt.dayofweek
    df['month'] = df['DATE'].dt.month
   
    return df
```
